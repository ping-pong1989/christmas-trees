```python
for i in range(1, 6):
    if i == 1:
        print(" "*(6-i), "*")
    elif i == 5:
        print(""*(6-1), "* "*i)
    else:
        print(" "*(6-i) + "* " + " "*(2*i - 2) + "*")
```
